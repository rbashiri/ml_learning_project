**Training Dataset**

```python
# import library
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
# load data 
df = pd.read_csv('/home/susan/machine_learning/Data/train_data_houses.csv')
# The following code allows us to display all the coulmns
pd.set_option('display.max_columns', None)
# Now you can see up to 100 rows at once
pd.set_option('display.width', None)
# Keep all rows, drop columns whose names start with "Unnamed"
#c stands for each column name
# it could be named anything, like col or column
df = df.drop(columns=[c for c in df.columns if c.startswith('Unnamed')])
print(df.shape)
print()
print(df.head())
print()
print(df.columns)
```

```python
# Many machine learning libraries require features to be stored in separate variables.
#  Initialize two variables:

features = df.drop(['last_price'], axis= 1)
target = df['last_price']
print(features.shape)
print()
print(target.shape)
```

**The train_test_split Function**

```python

# improt from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
print("Training set size:", X_train.shape[0])
print("Test set size:", X_test.shape[0])
print("Training features shape:", X_train.shape)
print("Test features shape:", X_test.shape)
```

```python
#Train Your Linear Regression Model
# Create a linear regression model, train it on your training data, 
# and make predictions on your test data
# Your task: Create and train the model
model = LinearRegression()
# Train the model
model.fit(X_train, y_train)
# Make predictions
predictions = model.predict(X_test)
mse = mean_squared_error(y_test, predictions)

# Your task: Calculate performance metrics
rmse = np.sqrt(mse)
print(f"Mean Squared Error: {mse:,.2f}")
print(f"Root Mean Squared Error: ${rmse:,.2f}")




```

**Train Your Logistic Regression Model**

- Create a logistic regression model, 
- train it on your training data
-  make predictions on your test data.
-  The data has already been loaded and split into a train and test set for you.



```python
# load data
diabet = pd.read_csv('/home/susan/machine_learning/Data/diabetes.csv')

# Many machine learning libraries require features to be stored in separate variables.
#  Initialize two variables:
diabet = diabet.drop(columns=["Unnamed: 0"], errors="ignore")
print(diabet.info())
print(diabet.head())
features = diabet.drop(['Outcome'], axis= 1)
target = diabet['Outcome']

# split data 
# improt from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
model = LogisticRegression(random_state=42,max_iter=100)
model.fit(X_train, y_train)
predictions = model.predict(X_test)
probability = model.predict_proba(X_test)
# Calculate metrics (assuming you have target_test and predictions)
accuracy = accuracy_score(y_test, predictions)
precision = precision_score(y_test, predictions)
recall = recall_score(y_test, predictions)
f1 = f1_score(y_test, predictions)
confusion_mat = confusion_matrix(y_test, predictions)
print("=== LOGISTIC REGRESSION PERFORMANCE ===")
print()
print(f"Accuracy:  {accuracy:.1%}")
print(f"Precision: {precision:.1%}")
print(f"Recall:    {recall:.1%}")
print(f"F1 Score:  {f1:.1%}")
print()
print("Confusion Matrix:")
print(confusion_matrix(y_test, predictions))
print()
print("Interpretation:")
print(f"• Model is correct {accuracy:.1%} of the time overall")
print(f"• When predicting diabetes, it's right {precision:.1%} of the time")
print(f"• Catches {recall:.1%} of all actual diabetes cases")
print(f"• F1 score of {f1:.1%} shows balanced performance")
```
